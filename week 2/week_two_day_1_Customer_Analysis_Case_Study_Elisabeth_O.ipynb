## insert data
- Aggregate data into one Data Frame using Pandas.
- Pay attention that files may have different names for the same column. therefore, make sure that you unify the columns names before concating them.

```python
def original_data():
    file1 = pd.read_csv("data/file1.csv")
    file2 = pd.read_csv("data/file2.csv")
    file3 = pd.read_csv("data/file3.csv")
    return pd.concat([file1,file2,file3], axis=0) #axis =0 means concat rowwise
```

```python
cu_df=original_data()
```

## Working with data types 
- Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of open complaints ). 
- remove the percentage from the customer lifetime value and truncate it to an integer value.

```python
cu_df.info()
```
