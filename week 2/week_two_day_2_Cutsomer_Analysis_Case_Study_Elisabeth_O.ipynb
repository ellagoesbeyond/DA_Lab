## Bucketing the data 
- Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

```python
state=cu_df["state"]
```

```python
# map(cu_df["state"])
def state_zones():
    cu_df.loc[cu_df["state"] == "California", "state"] = "West"
    cu_df.loc[cu_df["state"] == "Oregon", "state"] = "North West"
    cu_df.loc[cu_df["state"] == "Washington", "state"] = "East"
    cu_df.loc[(cu_df["state"] == "Arizona", "state")or(cu_df["state"] == "Nevada", "state")] = "Central"
    cu_df.rename(columns = {'state':'zones'})
    return 

    
```
